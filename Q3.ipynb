```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("Table-2019_58.csv")
```

```python
data
```

```python
oil = data['Oil']
natural_gas = data['Natural Gas']
coal = data['Coal']
nuclear_energy = data['Nuclear Energy']
hydro_electricity = data['Hydro electricity']
renewables = data['Renewables']
```

```python
plt.figure(figsize = (10, 5))
plt.scatter(data['Year'], data['Oil'], label = 'Oil')
plt.scatter(data['Year'], data['Natural Gas'], label = 'Natural Gas')
plt.scatter(data['Year'], data['Coal'], label = 'Coal')
plt.scatter(data['Year'], data['Nuclear Energy'], label = 'Nuclear Energy')
plt.scatter(data['Year'], data['Hydro electricity'], label = 'Hydro electricity')
plt.scatter(data['Year'], data['Renewables'], label = 'Renewables')
plt.legend(bbox_to_anchor = (1.05, 1) , loc = 'upper left')
plt.xlabel("Year")
plt.title(' Trends in World Primary Energy Consumption by Fuel from 2012 to 2018')
plt.ylabel("Consumption in million  tonnes of equivalent")
plt.show()
```

```python
complete_data = np.array([data['Oil'], data['Coal'], data['Natural Gas'], data['Nuclear Energy'], 
                          data['Hydro electricity'], data['Renewables']])
X = ['2012', '2013', '2014', '2015', '2016', '2017', '2018']
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(complete_data)
bar = np.arange(len(X))
plt.xticks(bar+1, X)
plt.xlabel('Year')
plt.ylabel('Consumption in million  tonnes of equivalent')
plt.title(' Trends in World Primary Energy Consumption by Fuel from 2012 to 2018')
plt.show()
```
